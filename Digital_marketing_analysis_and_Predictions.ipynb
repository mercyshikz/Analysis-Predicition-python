```python
#Defining the function and setting a parameter gender_column which represent Gender in my dataset.
def convert_to_buckets(gender_column):
#Calculate the gender counts.
    gender_counts = gender_column.value_counts()

#Initializing the bucket dictionary for the gender which are 2.
    buckets = {
        'Male': 0,
        'Female': 0,
    }
#Creating a loop to populate the Buckets over the gender_counts and then placing them in appropriate buckets as from the bucket dictionary.
    for gender, count in gender_counts.items():
        if gender == 'Male':
            buckets['Male'] += count
        else:
            buckets['Female'] += count
    return buckets

#Passing the Gender column to the function and Converting to buckets. 
bucketed_gender_counts = convert_to_buckets(DM_data['Gender'])

#Display the bucketed counts values.
print(bucketed_gender_counts)
```

```python
#Defining data to be used in the pie chart ploting.
labels = bucketed_gender_counts.keys() #This represents ~'female' and 'male'.
#Keys are extracted directly from the dictionary and used as labels to ensure the chart correctly reflects the dictionary structure.
sizes = bucketed_gender_counts.values() #This represents values~'female=4839' and 'male=3161'.
colors = ['#0000FF','#BE27C6']  #Colors for each slice. pink for female while blue for male.

#Plotting the pie chart.
plt.figure(figsize=(15, 8))
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=140,shadow=True,textprops={'fontsize': 14})

#Setting chart title.
plt.title('Gender Distribution',fontsize=16)

#Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal') 

#Displaying the chart.
plt.show()
```

```python
#defining the function and setting a parameter Age.
def convert_to_buckets(Age):
#Setting the ranges by using logical operates.
    if Age >= 0 and Age < 18: #This includes all ages from 0 to 17, but not 18. Since 17 is less than 18.
        return 'Youth(0-17)'
    elif Age >= 18 and Age < 26: #This includes all ages from 18 to 25, but not 26. Since 25 is less than 26.
        return 'Young Adult(18-25)' 
    elif Age >= 26 and Age < 45: #This includes all ages from 26 to 44, but not 45. Since 44 is less than 45.
        return 'Adult(26-44)'
    elif Age >= 45 and Age < 60: #This includes all ages from 45 to 59, but not 60. Since 59 is less than 60.
        return 'Middle Age(45-59)'
    elif Age >= 60 and Age <= 70: #This includes all ages from 60 to 70, Since 70 is less than and equal to our Age.
        return 'Old Age(60-70)'
    else: #This includes all ages values that fall outside the specified ranges the function returns 'Unknown' to indicating an undefined category.
        return 'Unknown'
        
#Applying the function to the 'Age' column and create a new column 'Age Group'
DM_data['Age Group'] = DM_data['Age'].apply(convert_to_buckets)

#Count the number of entries in each Age Group
age_group_counts = DM_data['Age Group'].value_counts()

#Display the counts for each Age Group
print(age_group_counts)     
```

```python
#Defining the function and setting a parameter income.
def convert_income_to_buckets(Income):
#Setting the ranges by using logical operates.
    if 20000 <= Income < 41000: #checks if Income is greater than or equal to $20,000 and less than $41,000,(ranges between $20,000 and $40,999).
        return 'Lower Income ($20,000-$40,999)'
    elif 41000 <= Income < 61000: #checks if Income is greater than or equal to $41,000 and less than $61,000,(ranges between $41,000 and $60,999).
        return 'Middle Lower Income ($41,000-$60,999)'
    elif 61000 <= Income < 81000: #checks if Income is greater than or equal to $61,000 and less than $81,000,(ranges between $61,000 and $80,999).
        return 'Upper Lower Income ($61,000-$80,999)'
    elif 81000 <= Income < 101000: #checks if Income is greater than or equal to $81,000 and less than $101,000,(ranges between $81,000 and $100,999).
        return 'Low Middle Income ($81,000-$100,999)'
    elif 101000 <= Income < 121000: #checks if Income is greater than or equal to $101,000 and less than $121,000,(ranges between $101,000 and $120,999).
        return 'Middle  Income ($101,000-$120,999)'
    elif 121000 <= Income < 141000: #checks if Income is greater than or equal to $121,000 and less than $141,000,(ranges between $121,000 and $140,999).
        return 'Middle Upper Income ($121,000-$140,999)'
    elif 141000 <= Income <= 155000: #checks if Income is greater than or equal to $141,000 and less than or equal to $155,000,(ranges between $141,000 and $155,000).
        return 'High Income ($141,000-$155,000)'
    else: #If Income does not fall into any of the ranges defined above that it is less than $20,000 or greater than $155,000 the function returns 'Unknown'.
        return 'Unknown' 

#Applying the function to the 'Income' column and create a new column 'Income Group'.
DM_data['Income Group'] = DM_data['Income'].apply(convert_income_to_buckets)

income_group_order = [
    'Lower Income ($20,000-$40,999)',
    'Middle Lower Income ($41,000-$60,999)',
    'Upper Lower Income ($61,000-$80,999)',
    'Low Middle Income ($81,000-$100,999)',
    'Middle  Income ($101,000-$120,999)',
    'Middle Upper Income ($121,000-$140,999)',
    'High Income ($141,000-$155,000)',
]

#Count the number of entries in each Income Group and Sorting value_counts() based on custom order.
income_group_counts = DM_data['Income Group'].value_counts().reindex(income_group_order)

#Display the counts for each Income Group.
print(income_group_counts)
```

**2. MARKETING SPECIFIC VARIABLES VISUALIZATION.**<br>
   + CampaignChannel:In the data marketing campaign is done by using Email, Social Media, SEO(Search Engine Optimization), PPC(Pay Per Click) and Referral.<br>
   + CampaignType:Type of the marketing campaign used are Awareness, Consideration, Conversion and Retention).<br>
   + AdSpend: Amount spent on the marketing campaign in USD.<br>
   + ClickThroughRate: Rate at which customers click on the marketing content.<br>
   + ConversionRate: Rate at which clicks convert to desired actions (e.g., purchases).<br>


```python
#Defining the function and setting a parameter campaignchannel_column which represent CampaignChannel in my dataset.
def convert_to_buckets(campaignchannel_column):
    
#This line uses value_counts() to calculate the frequency of each unique value in campaignchannel_column,
#and defining campaignchannel_counts before calling convert_to_buckets.  
    campaignchannel_counts = campaignchannel_column.value_counts()
    
#Intializing the bucket dictionary,creates a dictionary buckets with predefined keys types of campaign channels.
#Each key is initialized to 0,indicating that the bucket for each channel is empty at the start.
    buckets = {
        'Email': 0,
        'Social Media': 0,
        'Referral':0,
        'SEO': 0,
        'PPC': 0,
    }
#Creating a loop to populate the Buckets over the campaignchannel_counts and then placing them in appropriate buckets as from the bucket dictionary.
    for CampaignChannel, count in campaignchannel_counts.items():
        if CampaignChannel in buckets:
            buckets[CampaignChannel] += count
        else:
            buckets['Other'] += count

    return buckets

#Calling the convert_to_buckets function and passing the CampaignChannel column from a dataset called DM_data.
bucketed_campaignchannel_counts = convert_to_buckets(DM_data['CampaignChannel'])

#Displaying bucketed counts for each campaign channel.
print(bucketed_campaignchannel_counts)
```

```python
#Defining data to be used in the donut chart ploting.
labels = bucketed_campaignchannel_counts.keys() #This represents;email,social media,referral,SEO and PPC.
sizes = bucketed_campaignchannel_counts.values()  #This represents values.
colors = ['#800080', '#F08080', '#CCCCFF', '#CCCC00','#a04000'] #Colors for each slice. 

#Plotting the donut chart.
plt.figure(figsize=(15, 8)) #figure size.
wedges, texts, autotexts = plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True,
                                   wedgeprops={'edgecolor': 'black'},textprops={'fontsize': 14})

#Adding a circle at the center to make it a donut chart.
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

#Add a legend
plt.legend(wedges, labels,title="Campaign Channel", loc="best")

#Setting chart title.
plt.title('Campaign Channel Distribution',fontsize=16,pad=30)

#Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

#Displaying the chart.
plt.show()
```

<b>**EXPLANTION ON EVALUTION METRICS MEANING ABOVE:**<br></b>
+ The Mean Squared Error (MSE) is metric is used to measure the average of the squares of the errors between predicted values and actual values. It quantifies how much on average,the predictions deviate from the true values.
+ R-squared (R) metric indicates the proportion of variance in the target variable explained by the model. An R score closer to 1 means the model explains most of the variance.
+ Mean Absolute Error (MAE) metric gives the average of the absolute errors between actual and predicted values. Unlike MSE, it doesn't penalize large errors as heavily since it doesnt square the differences.
